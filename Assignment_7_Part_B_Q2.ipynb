```python
#What are the ten days with the highest average wind-speed at EWR?
import pandas as pd
df = pd.read_csv("weather.csv")
df2 = df[["origin","month","day","wind_speed"]]
df2 = df2[df.origin == 'EWR']
df2
```

```python
g = df2.groupby(['month','day'])
g
```

```python
for day,day_df in g:
    print(day)
    print(day_df)
```

```python
g2 = g.mean()
g2
```

```python
g3 = g2.nlargest(10,'wind_speed')
g3
```
