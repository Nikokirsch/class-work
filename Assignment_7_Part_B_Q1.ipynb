```python
#What is the min, max and mean temperatures for each month at EWR?
import pandas as pd
df = pd.read_csv('weather.csv')
df2 = df[["origin","month","temp"]]
df2 = df2[df.origin == 'EWR']
df2
```

```python
g = df2.groupby('month')
g
```

```python
for month,month_df in g:
    print(month)
    print(month_df)
```

```python
g2 = g.mean()
g2
```

```python
g3 = g.max()
g3
```

```python
g4 = g.min()
g4
```
